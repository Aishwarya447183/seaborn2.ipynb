```python
##Q1.Sure, here are five types of plots that can be created using the Seaborn library along with their uses:

Scatter Plot: A scatter plot displays the relationship between two variables by plotting them as points on a 2D plane. It is useful to identify patterns and relationships between the variables.

Histogram: A histogram is used to visualize the distribution of a single variable. It shows the frequency of values in different ranges or bins. It is useful to understand the central tendency, skewness, and spread of the data.

Box Plot: A box plot shows the distribution of data based on the quartiles and identifies the outliers. It is useful to detect and visualize the spread of data, outliers, and distribution.

Heatmap: A heatmap is a graphical representation of data using colors to visualize the values. It is useful to visualize the correlations between variables, and also useful in data exploration to identify patterns and trends.

Line Plot: A line plot displays the relationship between two variables as a continuous line. It is useful to display trends and changes over time. It is often used to visualize time series data.
```

```python
##Q2.Sure, here's an example of how to plot a line plot with timepoint on the x-axis and signal on the y-axis, with different lines for each combination of event and region in the fmri dataset:

import seaborn as sns
import matplotlib.pyplot as plt

# Load the fmri dataset
fmri_data = sns.load_dataset("fmri")

# Create a line plot with event and region as hue
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri_data)

# Add a title and labels
plt.title("FMRI Signal Intensity over Time")
plt.xlabel("Timepoint")
plt.ylabel("Signal Intensity")

# Show the plot
plt.show()
```

```python
##Q3.
import seaborn as sns
import matplotlib.pyplot as plt

# Load the titanic dataset
titanic_data = sns.load_dataset("titanic")

# Create a figure with two subplots
fig, axs = plt.subplots(ncols=2, figsize=(10, 5))

# Create a box plot of age by passenger class
sns.boxplot(x="pclass", y="age", data=titanic_data, ax=axs[0])
axs[0].set_title("Age by Passenger Class")

# Create a box plot of fare by passenger class
sns.boxplot(x="pclass", y="fare", data=titanic_data, ax=axs[1])
axs[1].set_title("Fare by Passenger Class")

# Adjust the spacing between the subplots
fig.tight_layout()

# Show the plot
plt.show()
```

```python
##Q4.

import seaborn as sns
import matplotlib.pyplot as plt

# Load the diamonds dataset
diamonds_data = sns.load_dataset("diamonds")

# Create a histogram of price with cut as hue
sns.histplot(data=diamonds_data, x="price", hue="cut")

# Add labels and title
plt.xlabel("Price")
plt.ylabel("Count")
plt.title("Distribution of Diamond Prices by Cut")

# Show the plot
plt.show()
```
